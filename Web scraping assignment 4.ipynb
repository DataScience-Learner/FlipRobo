```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd # Importing the required libraries
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains
```

```python
PATH = "C:\Program Files (x86)\chromedriver.exe" # Defining path
```

```python
driver = webdriver.Chrome(PATH)
driver.get("https://github.com/trending")

time.sleep(3) # Going to target link
```

```python
links = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a") # Finding required elements
```

```python
urls = [] 
```

```python
for i in links: # Getting URLs from the href attribute
    urls.append(i.get_attribute('href'))    
```

```python
titles = [] # Empty list
descriptions = []
cont_counts = []
langs = []
```

```python
for link in urls: # Going to each link and finding the follow data
    driver.get(link)
    time.sleep(2)
    
    try:
        element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "repository-container-header"))
        )
    except:
        driver.quit()
        
    try:
        desc = driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[1]/div/p')
        descriptions.append(desc.text) # Finding required elements
    except NoSuchElementException:
        descriptions.append('No description, website, or topics provided.')
        
    try:
        title = driver.find_element_by_xpath("//strong[@class='mr-2 flex-self-stretch']/a")
        titles.append(title.text) # Finding required elements
    except NoSuchElementException:
        titles.append('No title.')
        
    try:
        cont = driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[5]/div/h2/a/span')
        cont_counts.append(cont.text) # Finding required elements
    except NoSuchElementException:
        cont_counts.append('Solo contributor.')
        
    try:
        lang = driver.find_element_by_xpath("//a[@class='d-inline-flex flex-items-center flex-nowrap Link--secondary no-underline text-small mr-3']")
        langs.append(lang.get_attribute('textContent').replace('\n','')) # Finding required elements
    except NoSuchElementException:
        langs.append('Magic.')
    
```

```python
git_df = pd.DataFrame({}) # Creating data frame, and filling it with values we have scraped.
git_df['Repository tile'] = titles
git_df['Repository description'] = descriptions
git_df['Contributors count'] = cont_counts
git_df['Language used'] = langs
```

```python
git_df # View
```

```python
time.sleep(3)
driver.quit() # Quit
```
